```python
 #importing the pathlib and csv library
from pathlib import Path
```

```python
import csv
```

```python
#set the file path
csvpath = Path("C:/Users/Yvonne Colangelo/Downloads/budget_data.csv")
csv_output = Path("C:/Users/Yvonne Colangelo/python/FirstAssignment/python-homework/PyBank/analysis.txt")

#initialize a list that will hold months, revenue and a count variable for the number of months.
months=[]
revenue=[]
count = 0
total = 0
current_change = 0
changes=[]
greatest_profit = 0
greatest_loss = 0
new_count = 0
average_change = 0
   

    

#open the csv file as an object
#pass the csv file to the csv.reader() function
#(with ',' as the delimiter/separator) and return the csvreader object


with open (csvpath, "r") as csvfile:

    csvreader = csv.reader(csvfile, delimiter = ",")
    #initialize and assign the csv header. This will help not read the heading when iterating through the csvreader. 
    csv_header = next(csvreader)
       

   
    
    #Iterate through the rows in csvreader 
    #Assign the first row(list) to month and the second row(list) to profit_losses
    #Append list to month and revenue for iteration.
    
    
    for row in csvreader:
        #print(row)
        month = row[0]
        #print(month)
        profit_losses = int(row[1])
        months.append(month)
        revenue.append(profit_losses)
       
        # Get the length of month(list) for how many month are in the list
        count = len(months)
        
    #The next block of code is getting the average change in profit/losses over time              
    #To do this we are going to iterate through the revenue and getting the difference in revenue from month to month
    #In order to get the difference we assign the previous revenue 
    #We then iterate through revenue and get the differnce between the current profit or loss and the previous revenue on
    #We then append the list of differences to changes variable.
    
    previous_rev = revenue[0]
    for profit_loss in revenue:
        
        total +=profit_loss
        current_change = profit_loss - previous_rev
        previous_rev = profit_loss
        changes.append(current_change)
        
        
    #We then iterate through the list of changes adding them together and dividing them the number of months minus the first month
    #We round the final answer to the second decimal place.
    for change in changes:
        new_count += change
        average_change = round(new_count/(count-1), 2)
        
    
    #This block of code is for getting the greatest increase/decrease and the dates they occured
    #For this we get the maximum and minimum value in the list
    #We then use the index of the minimum/maximum value of the list to map out the index of the date in the months list. 
    greatest_profit = max(changes)
    greatest_loss = min(changes)
    
    index = changes.index(greatest_profit)
    #print(index)
    index_two = changes.index(greatest_loss)
   
    
    greatest_profit_date = months[index]
    greatest_loss_date = months[index_two]
   
    def final_results():
        
        print(f"Total Months : {count}")
        print(f"Total: {total}")
        print(f"Average Change : ${average_change}")  
        print(f"Greatest Increase in Profit: {greatest_profit_date} (${greatest_profit})")
        print(f"Greatest Decrease in Profits: {greatest_loss_date} (${greatest_loss})")
    
with open(csv_output, 'w') as txt_file:
    txt_file.write(f"Financial Analysis\n") 
    txt_file.write(f"Total Months : {count}\n")
    txt_file.write(f"Total: {total}\n")
    txt_file.write(f"Average Change : ${average_change}\n")
    txt_file.write(f"Greatest Increase in Profit: {greatest_profit_date} (${greatest_profit})\n")
    txt_file.write(f"Greatest Decrease in Profits: {greatest_loss_date} (${greatest_loss})\n")
    
    
        
    
```
